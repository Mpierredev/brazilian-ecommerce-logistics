```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv

# 1. Carrega as credenciais
load_dotenv()
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASS = os.getenv("DB_PASS")

DATABASE_URI = f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(DATABASE_URI)

# 2. Caminho de exportação para o Tableau
EXPORT_PATH = os.path.join("..", "data", "exports")
os.makedirs(EXPORT_PATH, exist_ok=True)

print("Conectando ao Data Warehouse e lendo a Camada Semântica...")

# 3. Consulta a View diretamente no banco
query = "SELECT * FROM vw_fato_vendas_completas;"
df_bi_feed = pd.read_sql(query, engine)

# 4. Exporta para CSV
file_name = "olist_bi_feed.csv"
file_path = os.path.join(EXPORT_PATH, file_name)

df_bi_feed.to_csv(file_path, index=False)

print(f"Base final para o Tableau gerada com {len(df_bi_feed)} linhas.")
print(f"Arquivo salvo em: {file_path}")
```
