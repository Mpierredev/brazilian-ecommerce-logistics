```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv

# 1. Carrega as variáveis de ambiente do arquivo .env
load_dotenv()

DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASS = os.getenv("DB_PASS")

# 2. Cria a string de conexão (Engine) do SQLAlchemy
DATABASE_URI = f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(DATABASE_URI)

# 3. Define os caminhos relativos
PROCESSED_DATA_PATH = os.path.join("..", "data", "processed")

# Mapeamento dos arquivos CSV para os nomes exatos das tabelas no PostgreSQL
tables_to_load = {
    "cleaned_customers.csv": "customers",
    "cleaned_sellers.csv": "sellers",
    "cleaned_products.csv": "products",
    "cleaned_geolocation.csv": "geolocation",
    "cleaned_category_translation.csv": "category_translation",
    "cleaned_orders.csv": "orders",
    "cleaned_items.csv": "order_items",
    "cleaned_payments.csv": "order_payments",
    "cleaned_reviews.csv": "order_reviews"
}

print("🚀 Iniciando ingestão automatizada no PostgreSQL...")

# 4. Loop de Ingestão
for file_name, table_name in tables_to_load.items():
    file_path = os.path.join(PROCESSED_DATA_PATH, file_name)
    
  
    df = pd.read_csv(file_path)
    
    df.to_sql(table_name, engine, if_exists='append', index=False)
    
    print(f"Tabela populada: {table_name} ({len(df)} linhas)")

print("\n Pipeline de Carga (Load) finalizado.")
```
